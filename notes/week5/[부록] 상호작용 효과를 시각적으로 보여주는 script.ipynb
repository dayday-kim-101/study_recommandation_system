# [부록]고객 메타정보들의 상호작용 효과를 시작적으로 확인해보기


### _Objective_
1. 고객이 상품의 구매에 영향을 미치는 다양한 변수가 서로 상호작용하는 경우를 가정하여 예측이 어떻게 달라지는지 확인하기

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def func1(x1,x2):
    return 0.5 + 0.3*x1 + 0.7*x2

x1s = np.linspace(1,5,20)
x2s = np.linspace(1,5,20)
points = np.stack(np.meshgrid(x1s,x2s),axis=-1)

ys = func1(points[...,0],points[...,1])
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
ax.scatter(points[...,0],points[...,1],ys)
ax.view_init(azim=-60)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.show()
```

```python
def func2(x1,x2):
    return 0.5 + -3.*(x1-2) + 1 * (x2-1) + 4. * x1 * x2
```

```python
x1s = np.linspace(-5,30,20)
x2s = np.linspace(-5,30,20)
points = np.stack(np.meshgrid(x1s,x2s),axis=-1)

ys = func2(points[...,0],points[...,1])

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
ax.scatter(points[...,0],points[...,1],ys)
ax.view_init(azim=-60)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
plt.show()
```
